### Explanation of the Code Components

#### **1. `\d+`**
- **Definition**: Matches one or more digits.
- **Example**: 
  - `'1234'`  Matches: `'1234'`.

---

#### **2. `\.?`**
- **Definition**: Matches an optional decimal point (`.`).
  - The `\` is used to escape the dot since its a special character in regex.
- **Example**: 
  - `'1234'`  Matches: No decimal point.
  - `'1234.'`  Matches: Decimal point.

---

#### **3. `\d+` (After the Decimal)**
- **Definition**: Matches one or more digits following the decimal point (if present).
- **Example**: 
  - `'1234.56'`  Matches: `'56'`.

---

#### **4. Behavior of `.findall()`**
- **Functionality**: Finds all matches of the regex pattern in a given string.
- **Returns**: A list of matches for each string in `X_res`.
- **Example**: 
  - Pattern: `\d+\.?\d*`
  - Input: `'1234.56'`
  - Output: `['1234.56']`

---

#### **5. `.apply(lambda x: x[0])`**
- **Functionality**: Applies a lambda function to each element of a column or series.
- **Purpose**: Extracts the first element of the list returned by `.findall()`.
- **Assumption**: Assumes theres at least one match in the list.
- **Example**: 
  - Input: `['1234.56']`
  - Output: `'1234.56'`

### Explanation of the Code Components

#### **1. `df['Cpu']`**
- **Definition**: Refers to the `Cpu` column in your DataFrame.
- **Purpose**: Accesses the data in the column for further processing.

---

#### **2. `.apply(lambda x: ...)`**
- **Definition**: Applies a function to each element of the `Cpu` column.
- **Explanation**: 
  - `x` represents a single entry (value) in the `Cpu` column.
  - The function defined within the `lambda` will be applied to every value in that column.

---

#### **3. `x.split()`**
- **Definition**: Splits the string `x` into a list of words based on spaces.
- **Example**:
  - Input: `'Intel Core i7'`
  - Output: `['Intel', 'Core', 'i7']`

---

#### **4. `x.split()[0:3]`**
- **Definition**: Extracts the first three words from the list created by `x.split()`.
- **Explanation**: 
  - `0:3` is a slice notation that picks the first three elements (index 0, 1, and 2).
  - Example:
    - Input: `'Intel Core i7 Processor'`
    - Output: `['Intel', 'Core', 'i7']`

---

#### **5. `" ".join(...)`**
- **Definition**: Joins the first three words (or any list of words) back into a single string with spaces between them.
- **Example**:
  - Input: `['Intel', 'Core', 'i7']`
  - Output: `'Intel Core i7'`

---

#### **6. `df['Cpu Name'] = ...`**
- **Definition**: Creates a new column in the DataFrame named `Cpu Name` to store the processed values.
- **Explanation**: 
  - The processed CPU names (first three words) are assigned to the new column `Cpu Name`.
  - Example: If the `Cpu` column contains values like `'Intel Core i7 Processor'`, the `Cpu Name` column will store `'Intel Core i7'`.

````python
# Step 1: Clean the 'Memory' column
df['Memory'] = df['Memory'].astype(str).replace(r'\.0', '', regex=True)
df["Memory"] = df["Memory"].str.replace('GB', '', regex=False)
df["Memory"] = df["Memory"].str.replace('TB', '000', regex=False)

# Step 2: Split into 'first' and 'second'
new = df["Memory"].str.split("+", n=1, expand=True)
df["first"] = new[0].str.strip()
df["second"] = new[1].str.strip().fillna("0")

# Step 3: Clean 'first' and 'second' to extract numeric values
df["first"] = df["first"].str.extract(r'(\d+)', expand=False).fillna("0")
df["second"] = df["second"].str.extract(r'(\d+)', expand=False).fillna("0")

# Step 4: Convert cleaned values to integers
df["first"] = df["first"].astype(int)
df["second"] = df["second"].astype(int)

# Step 5: Add storage type layers for 'first'
df["Layer1HDD"] = df["Memory"].apply(lambda x: 1 if "HDD" in x else 0)
df["Layer1SSD"] = df["Memory"].apply(lambda x: 1 if "SSD" in x else 0)
df["Layer1Hybrid"] = df["Memory"].apply(lambda x: 1 if "Hybrid" in x else 0)
df["Layer1Flash_Storage"] = df["Memory"].apply(lambda x: 1 if "Flash Storage" in x else 0)

# Step 6: Add storage type layers for 'second'
df["Layer2HDD"] = df["Memory"].apply(lambda x: 1 if "HDD" in x else 0)
df["Layer2SSD"] = df["Memory"].apply(lambda x: 1 if "SSD" in x else 0)
df["Layer2Hybrid"] = df["Memory"].apply(lambda x: 1 if "Hybrid" in x else 0)
df["Layer2Flash_Storage"] = df["Memory"].apply(lambda x: 1 if "Flash Storage" in x else 0)

# Step 7: Calculate total storage
df["HDD"] = (df["first"] * df["Layer1HDD"] + df["second"] * df["Layer2HDD"])
df["SSD"] = (df["first"] * df["Layer1SSD"] + df["second"] * df["Layer2SSD"])
df["Hybrid"] = (df["first"] * df["Layer1Hybrid"] + df["second"] * df["Layer2Hybrid"])
df["Flash_Storage"] = (df["first"] * df["Layer1Flash_Storage"] + df["second"] * df["Layer2Flash_Storage"])

# Step 8: Drop unnecessary columns
df.drop(columns=['first', 'second', 'Layer1HDD', 'Layer1SSD', 'Layer1Hybrid',
                 'Layer1Flash_Storage', 'Layer2HDD', 'Layer2SSD', 'Layer2Hybrid',
                 'Layer2Flash_Storage'], inplace=True)
````

## Step-by-Step Explanation in Simple Language

### Step 1: **Cleaning the 'Memory' Column**
```python
df['Memory'] = df['Memory'].astype(str).replace(r'\.0', '', regex=True)
df["Memory"] = df["Memory"].str.replace('GB', '', regex=False)
df["Memory"] = df["Memory"].str.replace('TB', '000', regex=False)
```

1. **Convert to String**: Ensure all values in the `'Memory'` column are strings (to avoid errors when cleaning).
2. **Remove `.0`**: Sometimes numbers like `128.0` appear. This removes `.0` from such cases.
3. **Remove 'GB'**: Since we only need the numbers, remove the "GB" label.
4. **Convert 'TB' to '000'**: `1TB` = `1000GB`. Replacing `TB` with `000` turns `1TB` into `1000`.

---

### Step 2: **Splitting the Memory Column**
```python
new = df["Memory"].str.split("+", n=1, expand=True)
df["first"] = new[0].str.strip()
df["second"] = new[1].str.strip().fillna("0")
```

1. **Split by `+`**: Some entries have two memory components like `"128GB + 1TB"`. We split them into two parts:
   - `first`  `"128GB"`
   - `second`  `"1TB"` (if available; otherwise `NaN`).
2. **Clean Whitespace**: Remove any spaces around the values.
3. **Handle Missing Values**: If `second` is empty (missing), replace it with `"0"`.

---

### Step 3: **Extract Only Numbers**
```python
df["first"] = df["first"].str.extract(r'(\d+)', expand=False).fillna("0")
df["second"] = df["second"].str.extract(r'(\d+)', expand=False).fillna("0")
```

1. **Extract Numbers**: For strings like `'128 SSD'`, we only take the numeric part (`128`) using a regular expression (`r'(\d+)'`).
   - `\d`  Matches digits.
   - `+`  Matches one or more digits.
2. **Fill Missing Numbers**: If no numbers are found, we replace them with `"0"`.

---

### Step 4: **Convert to Integers**
```python
df["first"] = df["first"].astype(int)
df["second"] = df["second"].astype(int)
```

1. **Convert Strings to Integers**: Now that we only have clean numbers, we convert them to integers for calculations.

---

### Step 5: **Identify Storage Types for Layer 1**
```python
df["Layer1HDD"] = df["Memory"].apply(lambda x: 1 if "HDD" in x else 0)
df["Layer1SSD"] = df["Memory"].apply(lambda x: 1 if "SSD" in x else 0)
df["Layer1Hybrid"] = df["Memory"].apply(lambda x: 1 if "Hybrid" in x else 0)
df["Layer1Flash_Storage"] = df["Memory"].apply(lambda x: 1 if "Flash Storage" in x else 0)
```

1. **Check for Each Type**: For each row in `'Memory'`, check if it contains a specific type of storage (e.g., `"HDD"`).
2. **Assign 1 or 0**:
   - If the storage type exists in the row, set it to `1` (true).
   - Otherwise, set it to `0` (false).

---

### Step 6: **Identify Storage Types for Layer 2**
```python
df["Layer2HDD"] = df["Memory"].apply(lambda x: 1 if "HDD" in x else 0)
df["Layer2SSD"] = df["Memory"].apply(lambda x: 1 if "SSD" in x else 0)
df["Layer2Hybrid"] = df["Memory"].apply(lambda x: 1 if "Hybrid" in x else 0)
df["Layer2Flash_Storage"] = df["Memory"].apply(lambda x: 1 if "Flash Storage" in x else 0)
```

This is similar to **Step 5**, but applies the same logic to the second memory layer.

---

### Step 7: **Calculate Total Storage**
```python
df["HDD"] = (df["first"] * df["Layer1HDD"] + df["second"] * df["Layer2HDD"])
df["SSD"] = (df["first"] * df["Layer1SSD"] + df["second"] * df["Layer2SSD"])
df["Hybrid"] = (df["first"] * df["Layer1Hybrid"] + df["second"] * df["Layer2Hybrid"])
df["Flash_Storage"] = (df["first"] * df["Layer1Flash_Storage"] + df["second"] * df["Layer2Flash_Storage"])
```

1. **Multiply Numbers with Flags**:
   - `df["first"] * df["Layer1HDD"]`: If the first layer is HDD, multiply the number in `first` with 1 (or 0 if it's not HDD).
   - Add the same for the second layer.
2. **Store Results**: For each storage type (HDD, SSD, Hybrid, Flash), store the total storage capacity.

---

### Step 8: **Drop Temporary Columns**
```python
df.drop(columns=['first', 'second', 'Layer1HDD', 'Layer1SSD', 'Layer1Hybrid',
                 'Layer1Flash_Storage', 'Layer2HDD', 'Layer2SSD', 'Layer2Hybrid',
                 'Layer2Flash_Storage'], inplace=True)
```

1. **Remove Unnecessary Columns**: After calculating the totals, we no longer need intermediate columns like `first`, `second`, or the layer flags.
2. **Keep Only Final Results**: The DataFrame now only contains the useful storage totals.

---

### Final Output
At the end of this process, your DataFrame will look something like this:

| Memory           | HDD   | SSD   | Hybrid | Flash_Storage |
|------------------|-------|-------|--------|---------------|
| 128GB + 1TB HDD  | 1024  | 128   | 0      | 0             |
| 256GB SSD        | 0     | 256   | 0      | 0             |
| 512GB Hybrid     | 0     | 0     | 512    | 0             |
| 128GB Flash      | 0     | 0     | 0      | 128           |

---

This code:
1. **Cleans up messy data** in the `'Memory'` column.
2. **Splits and processes storage layers** (primary and secondary).
3. **Identifies storage types** (HDD, SSD, etc.).
4. **Calculates total storage** for each type.

**Linear Regression**

````python
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder
from sklearn.linear_model import LinearRegression
from sklearn.pipeline import Pipeline
from sklearn.metrics import r2_score, mean_absolute_error
from sklearn.neighbors import KNeighborsRegressor
from sklearn.ensemble import ExtraTreesRegressor

step1 = ColumnTransformer(transformers=[
    ('col_tnf', OneHotEncoder(sparse_output=False, drop='first'), [0, 1, 7, 10, 11])
], remainder='passthrough')

# Step 2: Linear regression model
step2 = LinearRegression()

# Create pipeline: Combine both steps
pipe = Pipeline([
    ('step1', step1),
    ('step2', step2)
])

# Fit the pipeline to the training data
pipe.fit(X_train, y_train)

# Predict on test data
y_pred = pipe.predict(X_test)

# Calculate performance metrics
print('R2 score:', r2_score(y_test, y_pred))
print('MAE:', mean_absolute_error(y_test, y_pred))
````

***Extra Trees**

````python
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder
from sklearn.ensemble import ExtraTreesRegressor
from sklearn.pipeline import Pipeline
from sklearn.metrics import r2_score, mean_absolute_error


step1 = ColumnTransformer(
    transformers=[
        ('col_tnf', OneHotEncoder(sparse_output=False, drop='first'), [0, 1, 7, 10, 11])
    ],
    remainder='passthrough'  # Keep other columns unchanged
)


step2 = ExtraTreesRegressor(
    n_estimators=100,       
    random_state=3,         
    bootstrap=True,         
    max_samples=0.5,       
    max_features=0.75,      
    max_depth=15            
)


pipe = Pipeline([
    ('step1', step1),  
    ('step2', step2)   
])

# Fit the pipeline on training data
pipe.fit(X_train, y_train)

# Predict on the test data
y_pred = pipe.predict(X_test)

# Evaluate the model
print('R2 score:', r2_score(y_test, y_pred))  
print('MAE:', mean_absolute_error(y_test, y_pred)) 
````

***Stacking***

````python
from sklearn.ensemble import StackingRegressor
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder
from sklearn.ensemble import ExtraTreesRegressor
from sklearn.neighbors import KNeighborsRegressor
from sklearn.linear_model import LinearRegression, Ridge
from sklearn.pipeline import Pipeline
from sklearn.metrics import r2_score, mean_absolute_error


step1 = ColumnTransformer(transformers=[
    ('col_tnf', OneHotEncoder(sparse_output=False, drop='first'), [0, 1, 7, 10, 11])
], remainder='passthrough')

estimators = [
    ('et', ExtraTreesRegressor(n_estimators=100, random_state=3, bootstrap=True, 
                                max_samples=0.5, max_features=0.75, max_depth=15)),
    ('knn', KNeighborsRegressor(n_neighbors=3)),
    ('lr', LinearRegression())
]

step2 = StackingRegressor(estimators=estimators, final_estimator=Ridge(alpha=100))


pipe = Pipeline([
    ('step1', step1),  # Preprocessing step
    ('step2', step2)   # Model step
])

pipe.fit(X_train, y_train)


y_pred = pipe.predict(X_test)

print('R2 score:', r2_score(y_test, y_pred))
print('MAE:', mean_absolute_error(y_test, y_pred))
````
